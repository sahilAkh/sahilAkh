```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Given data
temperature_values = np.array([385, 455, 556, 714])  # in Kelvin
resistivity_values = np.array([0.028, 0.0061, 0.0013, 0.000274])  # in ohm-m

# Function to fit (Arrhenius equation)
def arrhenius_equation(T, sigma_0, Eg):
    k = 8.617333262145e-5  # Boltzmann constant in eV/K
    return sigma_0 * np.exp(-Eg / (2 * k * T))

# Fit the data to the Arrhenius equation
fit_params, _ = curve_fit(arrhenius_equation, temperature_values, 1 / resistivity_values)

# Extract the fitted parameters
sigma_0_fit, Eg_fit = fit_params

# Generate temperatures for the plot
temperature_plot = np.linspace(np.min(temperature_values), np.max(temperature_values), 1000)

# Calculate conductivity using the fitted parameters
conductivity_plot = arrhenius_equation(temperature_plot, sigma_0_fit, Eg_fit)

# Plot the data and the fitted curve
plt.scatter(temperature_values, 1 / resistivity_values, label='Experimental Data')
plt.plot(temperature_plot, conductivity_plot, label='Fitted Curve', color='red')
plt.xlabel('Temperature (K)')
plt.ylabel('Conductivity (S/m)')
plt.title('Conductivity vs Temperature with Arrhenius Fit')
plt.legend()
plt.show()

# Print the fitted parameters
print(f"Fitted sigma_0: {sigma_0_fit} S/m")
print(f"Fitted Eg: {Eg_fit} eV")
```

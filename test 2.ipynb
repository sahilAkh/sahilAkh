```python












from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have three different arrays for the pdf of test points of three separate classes
class1 = np.random.normal(0, 1, size=(100, 2))
class2 = np.random.normal(1, 1, size=(100, 2))
class3 = np.random.normal(2, 1, size=(100, 2))

# Combine the data and create labels
X = np.vstack([class1, class2, class3])
y = np.hstack([np.repeat(1, 100), np.repeat(2, 100), np.repeat(3, 100)])

# Fit the QDA model
model = QuadraticDiscriminantAnalysis()
model.fit(X, y)

# Create a grid to evaluate model
xx, yy = np.meshgrid(np.linspace(-5, 5, 500), np.linspace(-5, 5, 500))
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])
Z = Z[:, 1].reshape(xx.shape)

# Plot the decision boundaries
plt.contour(xx, yy, Z, [0.5], linewidths=2., colors='black')

# Plot the points
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
import numpy as np
from scipy.stats import multivariate_normal

def getArtificialData(mean, cov, nx, nt):
    # Generate data points
    x = np.random.multivariate_normal(mean, cov, nx)
    # Generate test points
    t = np.random.multivariate_normal(mean, cov, nt)
    # Compute mean based on x
    mu = np.mean(x, axis=0)
    # Compute covariance matrix based on x
    Sgm = np.cov(x, rowvar=False)
    return x, t, mu, Sgm

def evaluateMultiVarGauss(t, mu, Sgm):
    # Create a multivariate Gaussian object with specified mean and covariance matrix
    mvn = multivariate_normal(mu, Sgm)
    # Evaluate the probability density function at the points
    p = mvn.pdf(t)
    return p

def calculateTestSet(t, pXY, yEx):
    # Compute the labels
    y = np.argmax(pXY, axis=1)
    # Count the number of mislabelled points
    num_mislabelled = np.sum(y != yEx)
    # Print the probabilities and the actual and computed labels for each test point
    for i in range(len(t)):
        print(f"Point: {t[i]}, Probabilities: {pXY[i]}, Actual label: {yEx[i]}, Computed label: {y[i]}")
    print(f"Number of mislabelled points: {num_mislabelled}")

# Define the parameters for the three classes
params = [
    {"mean": [-0, 0], "cov": [[2, 1], [1, 50]], "nx": 40, "nt": 10},
    {"mean": [7, 5], "cov": [[3, 1], [1, 3]], "nx": 80, "nt": 20},
    {"mean": [-5, 5], "cov": [[5, 2], [2, 3]], "nx": 20, "nt": 5}
]

# Generate the data and test points for each class
data = [getArtificialData(**param) for param in params]

# Concatenate the test points of each class into one long vector
t = np.concatenate([d[1] for d in data])

# Evaluate the multivariate Gaussian for each class with the same merged test points
pXY = np.array([evaluateMultiVarGauss(t, *d[2:]) for d in data]).T

# Define the exact labels of the test points
yEx = np.concatenate([np.full(d[1].shape[0], i) for i, d in enumerate(data)])

# Calculate the test set
calculateTestSet(t, pXY, yEx)
```
